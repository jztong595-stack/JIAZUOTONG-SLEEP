```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from termcolor import colored

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import confusion_matrix
import shap

# ==================== 5. 读取数据 ====================
sleep_data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
print("数据形状:", sleep_data.shape)
sleep_data.head()

# ==================== 6. 数据预处理 ====================
# 6.1 血压二值化（基于医学标准）
def blood_pressure_label(bp_str):
    try:
        systolic, diastolic = map(int, bp_str.split('/'))
        return 0 if (systolic < 120 and diastolic < 80) else 1
    except:
        return 1

sleep_data['Blood Pressure'] = sleep_data['Blood Pressure'].apply(blood_pressure_label)

# 6.2 连续变量分箱
sleep_data["Age"] = pd.cut(sleep_data["Age"], 2)
sleep_data["Heart Rate"] = pd.cut(sleep_data["Heart Rate"], 4)
sleep_data["Daily Steps"] = pd.cut(sleep_data["Daily Steps"], 4)
sleep_data["Sleep Duration"] = pd.cut(sleep_data["Sleep Duration"], 3)
sleep_data["Physical Activity Level"] = pd.cut(sleep_data["Physical Activity Level"], 4)

# 6.3 标签编码
categorical_cols = ['Gender', 'Age', 'Occupation', 'Sleep Duration',
                    'Physical Activity Level', 'BMI Category', 'Heart Rate',
                    'Daily Steps', 'Sleep Disorder']
for col in categorical_cols:
    le = LabelEncoder()
    sleep_data[col] = le.fit_transform(sleep_data[col])

# 6.4 删除无用列
sleep_data.drop(['Person ID'], axis=1, inplace=True)

# ==================== 7. 特征相关性热力图 ====================
correlation = sleep_data.corr()
max_6_corr = correlation.nlargest(6, "Sleep Disorder")
plt.figure(figsize=(10, 8))
sns.heatmap(max_6_corr, annot=True, fmt=".2f", annot_kws={"size": 8},
            linewidths=0.5, cmap='BuPu')
plt.title('Top 6 features correlated with Sleep Disorder')
plt.savefig('corr_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== 8. 划分训练/测试集 ====================
X = sleep_data.iloc[:, :-1]
y = sleep_data.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=32, shuffle=True
)

print("训练集大小:", X_train.shape)
print("测试集大小:", X_test.shape)

# ==================== 9. 训练多个模型 ====================
models = {
    "Logistic Regression": LogisticRegression(),
    "XGBoost": XGBClassifier(),
    "CatBoost": CatBoostClassifier(verbose=False),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC()
}

predictions = {}
train_scores, test_scores = {}, {}

print("\n模型训练结果：")
for name, model in models.items():
    model.fit(X_train, y_train)
    train_scores[name] = round(model.score(X_train, y_train) * 100, 2)
    test_scores[name]  = round(model.score(X_test, y_test) * 100, 2)
    predictions[name] = model.predict(X_test)
    print(f"{name:20} -> 训练准确率: {train_scores[name]:5.2f}% | 测试准确率: {test_scores[name]:5.2f}%")

# ==================== 10. 混淆矩阵可视化 ====================
class_names = ['None', 'Sleep_Apnea', 'Insomnia']
plt.figure(figsize=(15, 7))

for i, (name, y_pred) in enumerate(predictions.items(), start=1):
    cm = confusion_matrix(y_test, y_pred)
    plt.subplot(2, 3, i)
    sns.heatmap(cm, annot=True, fmt='d', cmap='BuPu',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')

plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== 11. SHAP 模型解释 ====================
best_model = XGBClassifier()
best_model.fit(X_train, y_train)

explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test)

class_names = ['None', 'Sleep_Apnea', 'Insomnia']

# 判断 shap_values 的格式
if isinstance(shap_values, list):
    # 旧版：列表，每个元素形状 (n_samples, n_features)
    for i, class_name in enumerate(class_names):
        shap.summary_plot(shap_values[i], X_test, show=False)
        plt.title(f'SHAP Summary for {class_name}')
        plt.savefig(f'shap_{class_name}.png', dpi=300, bbox_inches='tight')
        plt.show()
else:
    # 新版：三维数组 (n_samples, n_features, n_classes)
    for i, class_name in enumerate(class_names):
        shap_values_class = shap_values[:, :, i]
        shap.summary_plot(shap_values_class, X_test, show=False)
        plt.title(f'SHAP Summary for {class_name}')
        plt.savefig(f'shap_{class_name}.png', dpi=300, bbox_inches='tight')
        plt.show()
```
